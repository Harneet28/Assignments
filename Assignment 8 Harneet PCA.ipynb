```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
df= pd.read_csv('Wine.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
x=df.iloc[:,1:]
y=df['Type']
```

```python
sc= StandardScaler()
x= sc.fit_transform(x)
x
```

```python
pca = PCA()
pca_values = pca.fit_transform(x)
pca_values
```

```python
pca.components_
```

```python
var = pca.explained_variance_ratio_
var
```

```python
Var = np.cumsum(np.round(var,decimals= 4)*100)
Var
```

```python
plt.plot(Var,color="blue");
```

```python
pca=PCA(n_components=3)
pcs=pca.fit_transform(x)
pcs
```

```python
data = pd.DataFrame(pcs,columns=['PC1','PC2','PC3'])
data
```

```python
data['Type']= y
data
```

```python
import seaborn as sns
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=data);
```

```python
sns.scatterplot(data=data, x='PC1', y='PC2', hue='Type');
```

```python
### K-Means Clustering
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS');
```

```python
clusters3=KMeans(3,random_state=30).fit(data)
clusters3
```

```python
clusters3.labels_
```

```python
data3=df.copy()
data3['clusters3id']=clusters3.labels_
data3
```
